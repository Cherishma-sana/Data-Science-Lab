```python
# 1. Text Preprocessing (Tokenization + Stop-word Removal)

import pandas as pd
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download required NLTK resources
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab', quiet=True)

# Sample movie description
text = "The movie was amazing, exciting, and full of action scenes!"

# Convert to lowercase
text = text.lower()

# Remove punctuation
text = text.translate(str.maketrans('', '', string.punctuation))

# Tokenization
try:
    tokens = word_tokenize(text)
except LookupError:
    print("punkt_tab not found. Using split() instead.")
    tokens = text.split()

# Stop-word removal
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words]

# Output
print("Original Text:")
print("The movie was amazing, exciting, and full of action scenes!")

print("\nTokens:")
print(tokens)

print("\nProcessed Tokens (Stop words removed):")
print(filtered_tokens)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

print("----- Content-Based Recommendation -----")

# Movie dataset
movies = pd.DataFrame({
    'Movie': [
        'Avengers',
        'Batman',
        'Superman',
        'Titanic',
        'Inception'
    ],
    'Description': [
        'action superhero marvel fight',
        'dark superhero action detective',
        'superhero flying action hero',
        'romance ship love drama',
        'dream science fiction thriller'
    ]
})

print("\nMovie Dataset:")
print(movies)

# TF-IDF
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(movies['Description'])

# Cosine Similarity
similarity = cosine_similarity(tfidf_matrix)

movie_name = "Batman"

movie_index = movies[movies["Movie"] == movie_name].index[0]

scores = list(enumerate(similarity[movie_index]))
scores = sorted(scores, key=lambda x: x[1], reverse=True)

print(f"\nRecommended Movies for {movie_name}:")

for i in scores[1:]:
    if i[1] > 0:
        print(f"{movies.iloc[i[0]]['Movie']}  (Similarity = {i[1]:.2f})")
```
