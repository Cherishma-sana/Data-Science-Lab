```python
import pandas as pd
import numpy as np

# Sample dataset
df = pd.DataFrame({
    'Age': [25, 30, np.nan, 40, 35],
    'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']
})

# Mean for numeric column
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Mode for categorical column
df['Department'] = df['Department'].fillna(df['Department'].mode()[0])

print(df)
```

```python
import pandas as pd
import numpy as np

# Sample dataset
df = pd.DataFrame({
    'Age': [25, 30, np.nan, 40, 35],
    'Department': ['HR', 'Finance', 'Finance', np.nan, 'IT']
})

# Fill missing values
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Department'] = df['Department'].fillna(df['Department'].mode()[0])

# Forward Fill
df_ffill = df.copy().ffill()
print(df_ffill)

# Backward Fill
df_bfill = df.copy().bfill()
print(df_bfill)
```

```python
# Deletion (drop rows or columns)

# Drop rows with missing values
df_drop_rows = df.dropna()
print("After dropping rows:\n", df_drop_rows)

# Drop columns with missing values
df_drop_cols = df.dropna(axis=1)
print("After dropping columns:\n", df_drop_cols)
```

```python
import pandas as pd

# Sample dataset with duplicates
df = pd.DataFrame({
    'ID': [1, 2, 2, 3, 4, 4],
    'Name': ['Alice', 'Bob', 'Bob', 'Charlie', 'David', 'David'],
    'Age': [25, 30, 30, 35, 40, 40],
    'Department': ['HR', 'IT', 'IT', 'Finance', 'Marketing', 'Marketing'],
    'Salary': [40000, 50000, 50000, 60000, 70000, 70000],
    'City': ['New York', 'Chicago', 'Chicago', 'Boston', 'Seattle', 'Seattle']
})

print("Original Data:\n")
print(df)

# Remove exact duplicates
df_exact = df.drop_duplicates()

print("\nAfter Exact Match Removal:\n")
print(df_exact)
```

```python

# Subset-Based Removal
# Remove duplicates based on selected key columns (e.g., ID or Name).
# python
# Remove duplicates based only on 'ID'
df_subset_id = df.drop_duplicates(subset=['ID'])
print("\nAfter Subset-Based Removal (ID):\n", df_subset_id)
# Remove duplicates based only on 'Name'
df_subset_name = df.drop_duplicates(subset=['Name'])
print("\nAfter Subset-Based Removal (Name):\n", df_subset_name)
```

```python

# Date Standardization (ISO 8601 format)
# python
import pandas as pd

# Sample dataset with inconsistent date formats
df = pd.DataFrame({
    'Date': ['2025-01-05', '05/01/2025', 'Jan 5, 2025', '2025.01.05']
})

# Convert all dates to standard ISO format (YYYY-MM-DD)
df['Date'] = pd.to_datetime(df['Date'], errors='coerce').dt.strftime('%Y-%m-%d')

print(df)
```

```python
import pandas as pd

# Sample dataset with inconsistent text cases
df = pd.DataFrame({
    'Name': ['Alice', 'BOB', 'charlie', 'DAVID', 'eVa', 'FRANK', 'Grace', 'hEnRy']
})

# Convert all names to lowercase
df['Name_lower'] = df['Name'].str.lower()

# Convert all names to uppercase
df['Name_upper'] = df['Name'].str.upper()

print(df)
```
